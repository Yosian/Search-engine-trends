```python
#importing all datasets from Zara and connecting them.
import numpy as np
import pandas as pd
zara1=pd.read_csv('/Users/myfirst.csv',header=0,sep=',')
```

```python
zara2=pd.read_csv('/Users/mysecond.csv',header=0,sep=',')
zara3=pd.read_csv('/Users/mysthird.csv',header=0,sep=',')
zara4=pd.read_csv('/Users/myfourth.csv',header=0,sep=',')
zara5=pd.read_csv('/Users/myfith.csv',header=0,sep=',')
zara6=pd.read_csv('/Users/mysixth.csv',header=0,sep=',')
zara7=pd.read_csv('/Users/myseventh.csv',header=0,sep=',')
zara8=pd.read_csv('/Users/myeigth.csv',header=0,sep=',')
zara9=pd.read_csv('/Users/my9.csv',header=0,sep=',')
zara10=pd.read_csv('/Users/my10.csv',header=0,sep=',')
zara11=pd.read_csv('/Users/my11.csv',header=0,sep=',')
zara12=pd.read_csv('/Users/my12.csv',header=0,sep=',')
```

```python
frames=[zara1,zara2,zara3,zara4,zara5,zara6,zara7,zara8,zara9,zara10,zara11,zara12]
zara=pd.concat(frames)
```

```python
zara['date']=zara['date'].map(lambda x: x.rstrip('00:00:00+00:00'))
```

```python
zara.dropna(axis=0,how='any',inplace=True)
```

```python
del frames
del zara1,zara2,zara3,zara4,zara5,zara6,zara7,zara8,zara9,zara10,zara11,zara12
```

```python
zara.groupby('query').count().sort_values(['count'],ascending=False)
```

```python
zarapumps=zara[zara['query']=='pumps'].sort_values(['date'],ascending=True)#let us start with a popular item
```

```python
zarapumps=zarapumps.rename(columns={'date':'ds','count':'y'})
```

```python
zarapumps['ds']=pd.to_datetime(zarapumps['ds'])
```

```python
import fbprophet
import matplotlib.pyplot as plt
```

```python
plt.plot(zarapumps['ds'],zarapumps['y'])
```

```python
zarapumps_prophet=fbprophet.Prophet(changepoint_prior_scale=0.05)
zarapumps_prophet.fit(zarapumps)
```

```python
zarapumps_forecast=zarapumps_prophet.make_future_dataframe(periods=60,freq='D')
zarapumps_forecast=zarapumps_prophet.predict(zarapumps_forecast)
```

```python
zarapumps_prophet.plot(zarapumps_forecast,xlabel='Timeline',ylabel='Counts')
plt.title('Search of the item=PUMPS over 2017')
```

```python
zarapumps_prophet.plot_components(zarapumps_forecast)
plt.title('Trends by frequency')
```
